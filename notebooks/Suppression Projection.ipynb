```python
import sys
sys.path.insert(0, "../")
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
import pickle
window_size =200

with open(f"../runtime_generated/suppressed_models/scores_dict_{window_size}.pkl", 'rb') as handle:
    scores_dict = pickle.load(handle)
```

```python
fig, ax = plt.subplots(1, 1, figsize = (20, 10))

for key in scores_dict:
    train_scores, val_scores = scores_dict[key]
    
    train_err, train_loss = map(list, zip(*train_scores))
    val_err, val_loss = map(list, zip(*val_scores))

    ax.plot(1- np.array(train_err), label = "Train: " + str(key))
    ax.plot(1- np.array(val_err), label = "Test: " + str(key))
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Accuracy")

ax.legend()

```
